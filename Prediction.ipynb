## Predicting Life Expectancy with PyCaret

*This notebook provides a step-by-step guide on how to use PyCaret, an open-source, low-code machine learning library, to predict life expectancy at birth for a given population. The data used in this example is sourced from the World Bank.*

### Importing Libraries and Data

First, we import the necessary libraries and load the two datasets. The first dataset contains the target variable, which is the life expectancy at birth, and some features that may be used to predict it. The second dataset contains similar features but no target variable.

```python
import pandas as pd
import numpy as np
from pycaret.regression import *
from pycaret.utils import check_metric

df1 = pd.read_csv(r'C:\Users\Asus\Desktop\Project\Estimates Output.csv')

df2 = pd.read_csv(r'C:\Users\Asus\Desktop\Project\Medium variant Output.csv')
```

```python
df1.head()
```

```python
df2.head()
```

### Data Preparation

We rename the target column in both datasets to 'target' to ensure consistency, and drop the target variable from the second dataset since it will not be used in the training of the model.

```python
df1 = df1.rename(columns={'Life Expectancy at Birth, both sexes (years)_transformed': 'target'})

df2 = df2.rename(columns={'Life Expectancy at Birth, both sexes (years)_transformed': 'target'})

y = df2.target

del df2['target']
```

We then check the correlation between the features in the first dataset, and drop any features that are highly correlated with each other (correlation coefficient > 0.90).

```python
corr_matrix = df1.corr()
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
to_drop = [column for column in upper_tri.columns if any(abs(upper_tri[column]) > 0.90)]
df1.drop(to_drop, axis=1, inplace=True)
```

```python
df1.shape
```

Next, we ensure that the features in the second dataset are in the same order as the features in the first dataset.

```python
ls = df1.columns.values.tolist()
df2 = df2.filter(ls)
```

```python
df2.shape
```

We split the first dataset into two: one for modeling (90%) and another for prediction (10%).



```python
data = df1.sample(frac=0.9, random_state=789)
data_unseen = df1.drop(data.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions ' + str(data_unseen.shape))
```

### Model Training

We use PyCaret's setup function to initialize the training environment. We pass the modeling data and target variable to the function, as well as a session ID for reproducibility.

```python
reg = setup(data = df1, target = 'target', session_id = 123, silent = True)
```

Next, we use the compare_models function to compare the performance of all available regression models in PyCaret, excluding the ransac algorithm.

```python
best = compare_models(exclude = ['ransac'], fold = 5)
```

We select the Extra Trees Regressor (ET) as the best performing model.

```python
et = create_model('et', fold = 5)
```

This function prints the details of the Extra Trees model.

```python
print(et)
```

This function generates a plot of the Extra Trees model's performance using a number of evaluation metrics.

```python
plot_model(et)
```

This function generates a plot of the Extra Trees model's feature importance.

```python
plot_model(et, plot = 'feature')
```

This function creates an ensemble model by bagging the Extra Trees model with 5-fold cross-validation.

```python
bagged_et = ensemble_model(et, fold = 5)
```

We generate a plot of the bagged Extra Trees model performance using a number of evaluation metrics.

```python
plot_model(bagged_et)
```

This function finalizes the bagged Extra Trees model by training it on the entire dataset.

```python
final_bagged_et = finalize_model(bagged_et)
```

This function prints the details of the finalized bagged Extra Trees model.

```python
print(final_bagged_et)
```

We generate a plot of the finalized bagged Extra Trees model's residuals.

```python
plot_model(final_bagged_et, plot = 'residuals')
```

This function predicts the target variable for the training dataset using the finalized bagged Extra Trees model.

```python
predict_model(final_bagged_et);
```

After finalizing the model, we can use it to make predictions on the unseen data (data_unseen) using the predict_model function. The resulting predictions are stored in unseen_predictions. We also use the check_metric function to evaluate the R-squared score of the model on the unseen data.

```python
unseen_predictions = predict_model(final_bagged_et, data=data_unseen)
unseen_predictions.head()
check_metric(unseen_predictions.target, unseen_predictions.Label, 'R2')
```

Finally, we save the final model using the save_model function and load it back using the load_model function. We can then use the loaded model to make predictions on new data (df2) using the predict_model function. We again use the check_metric function to evaluate the R-squared score of the model on the new data.

```python
save_model(final_bagged_et, 'Final Bagged Et Model')
```

```python
saved_final_bagged_et = load_model('Final Bagged Et Model')
```

```python
new_prediction = predict_model(saved_final_bagged_et, data=df2)
```

```python
new_prediction.head()
```

```python
from pycaret.utils import check_metric
check_metric(y, new_prediction.Label, 'R2')
```

The predicted R-squared (R2) value of 0.9526 for the new data suggests that the finalized bagged Extra Trees model is a good fit for the new dataset. R2 is a statistical measure that represents the proportion of the variance in the dependent variable that is predictable from the independent variables. The R2 value ranges between 0 and 1, where 1 indicates a perfect fit between the predicted and actual values.

An R2 value of 0.9526 suggests that the model can explain 95.26% of the variance in the dependent variable based on the independent variables. This is a high value, which indicates that the model is performing well on the new dataset. However, it is important to note that the performance of the model on the new dataset should be evaluated using multiple evaluation metrics and not just R2 to ensure that the model is making accurate predictions.
